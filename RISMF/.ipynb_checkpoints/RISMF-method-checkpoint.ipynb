```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# input : S the sampling operator, M the list of values corresponding to S
def RISMF(S,M,sizeM,learningRate = 0.1, regularizedFactor = 0.1 , K = 1, percentageTrainingSet = 0.1) :
    #separate the sampling operator into training (W1) and validation (W2) set
#     W1 = S[0:(1-percentageTrainingSet)*len(S),:]
#     W2 = S[(1-percentageTrainingSet)*len(S):len(S),:]
    W1 = S
    W2 = []
    #initialize randomly P and Q in interval [-0.01,0.01]
    P = np.random.rand(sizeM[0], K)/50 - 0.01
    Q = np.random.rand(K,sizeM[1])/50 - 0.01
    Popt = np.zeros(shape=P.shape)
    Qopt = np.zeros(shape=Q.shape)
    #initialize the minimum RMSE
    minRMSE = float('inf')
    gradEp = np.zeros(K)
    gradEq = np.zeros(K)
    #initialize variables for stopping algorithm
    notDecreaseRMSE = 0
    cond = True
    while(cond) :
        k=0
        for(u,i) in W1 :
            e = M[k]-np.dot(P[u,:],Q[:,i])
            ePrime = (e**2 + regularizedFactor*P[u,:]*P[u,:].T + regularizedFactor*Q[:,i].T*Q[:,i])/2
            gradEp = -e*Q[:,i]+regularizedFactor*P[u,:]
            gradEq = -e*P[u,:]+regularizedFactor*Q[:,i]
            P[u,:] = P[u,:] - learningRate*gradEp
            Q[:,i] = Q[:,i] - learningRate*gradEq
            k=k+1
        #compare RMSE of the validation set W2
        currentRMSE = RMSE(M,W2,W1,P,Q,regularizedFactor)
        print(currentRMSE)
        if(currentRMSE<minRMSE) :
            minRMSE = currentRMSE
            Popt = P
            Qopt = Q
            notDecreaseRMSE = 0
        else :
            notDecreaseRMSE = notDecreaseRMSE + 1
        cond = notDecreaseRMSE!=5
    res = np.dot(Popt,Qopt)
    return res, Popt, Qopt

def RMSE(M,W1,W2,P,Q,regularizedFactor) :
    sse = 0
    size = len(W1)
    for (u,i) in W2 :
        e = M[size]-np.dot(P[u,:],Q[:,i])
        sse = sse + (0.5)*(e**2 + regularizedFactor*np.dot(P[u,:],P[u,:].T)  + regularizedFactor*np.dot(Q[:,i].T,Q[:,i]))
        size = size + 1
    return np.sqrt(sse/size)
```

```python
def sampling_operator(nsamp, dims, seed=1):
    np.random.seed(seed)
    idx = np.random.choice(np.prod(dims), nsamp, replace=False)
    
    return np.vstack(np.unravel_index(idx, dims)).T    

def get_matrix(rank, dims, seed=1):
    np.random.seed(seed)
    M_l = np.random.randn(dims[0], rank)
    M_r = np.random.randn(dims[1], rank)
    
    return M_l.dot(M_r.T)
```

```python
#learning rate, regularized factor and nb K
learningRate = 0.1
regularizedFactor = 0.1
K = 1
sizeM = (1000,1000)

S = sampling_operator(1000, sizeM)
MtoGuess = get_matrix(10, sizeM)
M = []
for (i,j) in S :
    M.append(MtoGuess[i,j])

#P,Q, rmse = RISMF(S,M,sizeM,learningRate, regularizedFactor, K)
#print(rmse)
percentage = 0.5
res, P, Q = RISMF(S,M,sizeM,learningRate, regularizedFactor, K, percentage)
print(RMSE(M,[],S,P,Q,regularizedFactor))
```

```python
learningRate = 0.01
bestLearningRate = learningRate
regularizedFactor = 0.01
bestRegularizedFactor = regularizedFactor
bestRMSE = float("inf")
for rate in range(10) :
    regularizedFactor = 0.01
    for factor in range(10) :
        P,Q, rmse = RISMF(S,M,sizeM,learningRate, regularizedFactor, K)
        if(rmse < bestRMSE) :
            bestRMSE = rmse
            bestLearningRate = learningRate
            bestRegularizedFactor = regularizedFactor
        regularizedFactor = regularizedFactor+0.01
    learningRate = learningRate+0.01
print("bestRMSE =",bestRMSE)
print("bestLearningRate =", bestLearningRate)
print("bestRegularizedFactor =", bestRegularizedFactor)
```

```python
lala = []
bestRMSE = float("inf")
bestK = 0
for K in range(1,50) :
    P,Q, rmse = RISMF(S,M,sizeM,bestLearningRate, bestRegularizedFactor, K)
    lala.append(rmse)
    if(rmse < bestRMSE) :
        bestRMSE = rmse
        bestK = K
print("bestRMSE = ", bestRMSE)
print("bestK = ", bestK)
```
